```python
event_max_losing_score = {}
for event in ['2026oncmp1', '2026oncmp2']:
    max_losing_score = 0
    for match in tba.event_matches(event):
        if match['comp_level'] != 'qm':
            continue

        blue_score = match['alliances']['blue']['score']
        red_score = match['alliances']['red']['score']

        if blue_score == red_score:
            losing_score = 0
        elif match['winning_alliance'] == 'blue':
            losing_score = red_score
        else:
            losing_score = blue_score

        max_losing_score = max(max_losing_score, losing_score)

    event_max_losing_score[event] = max_losing_score

print('Max losing score for qm matches per event:', event_max_losing_score)
```

```python
def extract_match_number(match_key):
    """
    Extract the match number from a match key.
    Example: '2026oncmp2_qm10' -> 10
    """
    parts = match_key.split('_qm')
    if len(parts) == 2:
        return int(parts[1])
    return 0


def allocate_nocolour_teams(event_key, blueteams, redteams, nocolourteams):
    """
    Allocate nocolourteams to blue or red based on their QM teammates.
    Returns a dict mapping team number to allocated color ('blue' or 'red')
    """
    allocation = {}
    
    for team in nocolourteams:
        team_key = f'frc{team}'
        team_matches = []
        
        # Get all QM matches for this team
        for match in tba.event_matches(event_key):
            if match['comp_level'] != 'qm':
                continue
            
            if team_key in match['alliances']['blue']['team_keys'] or team_key in match['alliances']['red']['team_keys']:
                match_number = extract_match_number(match['key'])
                team_matches.append((match_number, match))
        
        # If no matches found, skip
        if not team_matches:
            continue
        
        # Find teammates in all matches and their colors
        has_blue_teammate = False
        has_red_teammate = False
        
        for match_number, match in team_matches:
            # Get all teammates
            if team_key in match['alliances']['blue']['team_keys']:
                blue_teammates = [t for t in match['alliances']['blue']['team_keys'] if t != team_key]
                red_teammates = match['alliances']['red']['team_keys']
            else:  # team is in red
                red_teammates = [t for t in match['alliances']['red']['team_keys'] if t != team_key]
                blue_teammates = match['alliances']['blue']['team_keys']
            
            # Check teammate colors
            for teammate in blue_teammates:
                teammate_num = int(teammate[3:])  # Extract number from 'frcXXXX'
                if teammate_num in blueteams:
                    has_blue_teammate = True
                elif teammate_num in redteams:
                    has_red_teammate = True
            
            for teammate in red_teammates:
                teammate_num = int(teammate[3:])
                if teammate_num in blueteams:
                    has_blue_teammate = True
                elif teammate_num in redteams:
                    has_red_teammate = True
        
        # Allocate based on rules
        if has_blue_teammate and has_red_teammate:
            # Both colors - allocate based on LAST match (highest match number)
            last_match_number, last_match = max(team_matches, key=lambda x: x[0])
            if team_key in last_match['alliances']['blue']['team_keys']:
                allocation[team] = 'blue'
            else:
                allocation[team] = 'red'
        elif has_blue_teammate:
            allocation[team] = 'blue'
        elif has_red_teammate:
            allocation[team] = 'red'
        # If neither (shouldn't happen), leave unallocated
    
    return allocation


def calculate_points_by_colour(event_key, blueteams, redteams, nocolourteams):
    """
    Calculate total fuel scored by blue and red alliances in QM matches.
    Allocates nocolourteams appropriately.
    
    Returns detailed statistics including total fuel and team-level breakdowns.
    """
    # Get allocation for nocolourteams
    allocation = allocate_nocolour_teams(event_key, blueteams, redteams, nocolourteams)
    
    blue_total = 0
    red_total = 0
    blue_matches = 0
    red_matches = 0
    
    # Go through all QM matches
    for match in tba.event_matches(event_key):
        if match['comp_level'] != 'qm':
            continue
        
        blue_fuel = match['score_breakdown']['blue']['hubScore']['totalCount']
        red_fuel = match['score_breakdown']['red']['hubScore']['totalCount']
        
        blue_total += blue_fuel
        red_total += red_fuel
        blue_matches += 1
        red_matches += 1
    
    return {
        'blue_fuel': blue_total,
        'red_fuel': red_total,
        'blue_matches': blue_matches,
        'red_matches': red_matches,
        'blue_avg': blue_total / blue_matches if blue_matches > 0 else 0,
        'red_avg': red_total / red_matches if red_matches > 0 else 0,
        'team_allocations': allocation
    }
```

```python

def calculate_points_per_team(event_key, blueteams, redteams, nocolourteams):
    """
    Calculate fuel per team, with nocolourteams allocated appropriately.
    Returns a dataframe with team number, allocated colour, and total fuel from their alliance.
    """
    # Get allocation for nocolourteams
    allocation = allocate_nocolour_teams(event_key, blueteams, redteams, nocolourteams)
    
    team_data = []
    
    # Process each QM match
    for match in tba.event_matches(event_key):
        if match['comp_level'] != 'qm':
            continue
        
        blue_fuel = match['score_breakdown']['blue']['hubScore']['totalCount']
        red_fuel = match['score_breakdown']['red']['hubScore']['totalCount']
        
        # Blue teams
        for team_key in match['alliances']['blue']['team_keys']:
            team_num = int(team_key[3:])
            
            # Determine colour for this team
            if team_num in blueteams:
                colour = 'blue'
            elif team_num in redteams:
                colour = 'red'
            elif team_num in allocation:
                colour = allocation[team_num]
            else:
                colour = 'unknown'
            
            team_data.append({
                'team': team_num,
                'match': match['key'],
                'colour': colour,
                'alliance_colour': 'blue',
                'alliance_fuel': blue_fuel
            })
        
        # Red teams
        for team_key in match['alliances']['red']['team_keys']:
            team_num = int(team_key[3:])
            
            # Determine colour for this team
            if team_num in blueteams:
                colour = 'blue'
            elif team_num in redteams:
                colour = 'red'
            elif team_num in allocation:
                colour = allocation[team_num]
            else:
                colour = 'unknown'
            
            team_data.append({
                'team': team_num,
                'match': match['key'],
                'colour': colour,
                'alliance_colour': 'red',
                'alliance_fuel': red_fuel
            })
    
    df = pd.DataFrame(team_data)
    
    # Calculate totals per team
    team_totals = df.groupby('team').agg({
        'alliance_fuel': 'sum',
        'colour': 'first',  # All rows for a team have same colour
        'match': 'count'     # Number of matches played
    }).rename(columns={'match': 'matches_played'}).reset_index()
    
    team_totals['avg_fuel_per_match'] = team_totals['alliance_fuel'] / team_totals['matches_played']
    team_totals = team_totals.sort_values('alliance_fuel', ascending=False)
    
    return team_totals, df


# Calculate per-team statistics
team_totals, team_match_df = calculate_points_per_team(event, blueteams, redteams, nocolourteams)

print("\n" + "="*80)
print("FUEL PER TEAM BREAKDOWN")
print("="*80)

print("\nTop 20 Teams by Total Alliance Fuel:")
print(team_totals[['team', 'colour', 'alliance_fuel', 'matches_played', 'avg_fuel_per_match']].head(20).to_string(index=False))

# Summary by colour
print("\n" + "-"*80)
print("SUMMARY BY ALLOCATED COLOUR:")
print("-"*80)
for colour in ['blue', 'red']:
    colour_teams = team_totals[team_totals['colour'] == colour]
    total_fuel = colour_teams['alliance_fuel'].sum()
    avg_per_team = colour_teams['alliance_fuel'].mean()
    num_teams = len(colour_teams)
    
    print(f"\n{colour.upper()} Teams:")
    print(f"  Number of teams: {num_teams}")
    print(f"  Total alliance fuel: {total_fuel:,}")
    print(f"  Average fuel per team: {avg_per_team:.1f}")
```
