```python
list1 = []
for j in ['2024oncmp1','2024oncmp2']:
    for i in tba.event_matches(j):
        if i['comp_level'] == 'qm' and i['winning_alliance'] in ('blue', 'red'):
            if i['winning_alliance'] == 'blue':
                lose = 'red'
            else:
                lose = 'blue'
            score = i['alliances'][lose]['score']
            key = i['key']
            
            list1.append({key:score})
            
counter = collections.Counter()
for d in list1:
    counter.update(d)

res = dict(sorted(counter.items(), key=lambda item: item[1], reverse=True))

print('Highest Losing Score')
res
```

```python
penalty = 0
total = 0
end = 0
for j in ['2024oncmp1','2024oncmp2']:
    for i in tba.event_matches(j):
        if (i['comp_level'] == 'qm' and i['actual_time'] is not None):
            blue_end = i['score_breakdown']['blue']['melodyBonusAchieved']*1
            red_end = i['score_breakdown']['red']['melodyBonusAchieved']*1
            end += blue_end + red_end
            total += 2

print('Melodies:', end)
print('Total opportunities:', total)
print('Percentage: ', end/total)
```

```python
penalty = 0
total = 0
end = 0
for j in ['2024oncmp1','2024oncmp2']:
    for i in tba.event_matches(j):
        if (i['comp_level'] == 'qm' and i['actual_time'] is not None):
            blue = i['score_breakdown']['blue']['g424Penalty']*1
            red = i['score_breakdown']['red']['g424Penalty']*1
            blue_end = i['score_breakdown']['blue']['ensembleBonusAchieved']*1
            red_end = i['score_breakdown']['red']['ensembleBonusAchieved']*1
            penalty += blue + red
            end += blue_end + red_end
            total += 2

print('G424 violations:', penalty)
print('Ensembles:', end)
print('Real Ensembles', end-penalty)
print('Total opportunities:', total)
print('Ratio: ', penalty/(end-penalty), ':1')
```

```python
symp = 0
for j in ['2024oncmp1','2024oncmp2']:
    for i in tba.event_matches(j):
        for c in ['red', 'blue']:
            if i['comp_level'] == 'qm':

                harm = 1*(i['score_breakdown'][c]['endGameHarmonyPoints']>0)
                ens = 1*(i['score_breakdown'][c]['ensembleBonusAchieved'])
                mel = 1*(i['score_breakdown'][c]['melodyBonusAchieved'])
                spot = 1*(i['score_breakdown'][c]['endGameSpotLightBonusPoints']>0)
                trap = 1*(i['score_breakdown'][c]['endGameNoteInTrapPoints']>0)
                coop = 1*(i['score_breakdown'][c]['coopertitionBonusAchieved'])
                if harm + ens + mel + spot + trap + coop >= 5:
                    print(i['key'], harm + ens + mel + spot + trap + coop)
                
                symp += 1*((harm + ens + mel + spot + trap + coop) == 6)

print('Perfect Symphonies')
print(symp)
```

```python
def endgame_pts(value):
    if value == 'Parked': 
        return 1
    elif value in ('StageLeft', 'CenterStage', 'StageRight'): 
        return 3
    else: 
        return 0

list1 = []
for i in tba.event_matches(event):
    if i['comp_level'] == 'qm':
        teamb1 = i['alliances']['blue']['team_keys'][0]
        teamb2 = i['alliances']['blue']['team_keys'][1]
        teamb3 = i['alliances']['blue']['team_keys'][2]
        teamr1 = i['alliances']['red']['team_keys'][0]
        teamr2 = i['alliances']['red']['team_keys'][1]
        teamr3 = i['alliances']['red']['team_keys'][2]
        
        endb1 = endgame_pts(i['score_breakdown']['blue']['endGameRobot1'])
        endb2 = endgame_pts(i['score_breakdown']['blue']['endGameRobot2'])
        endb3 = endgame_pts(i['score_breakdown']['blue']['endGameRobot3'])
        endr1 = endgame_pts(i['score_breakdown']['red']['endGameRobot1'])
        endr2 = endgame_pts(i['score_breakdown']['red']['endGameRobot2'])
        endr3 = endgame_pts(i['score_breakdown']['red']['endGameRobot3'])
        
        dict1 = { teamb1:endb1
                , teamb2:endb2
                , teamb3:endb3
                , teamr1:endr1
                , teamr2:endr2
                , teamr3:endr3}
        list1.append(dict1)

counter = collections.Counter()
for d in list1:
    counter.update(d)
    
res = dict(sorted(counter.items(), key=lambda item: item[1], reverse=True))

print('End Game Pts')
res
```
