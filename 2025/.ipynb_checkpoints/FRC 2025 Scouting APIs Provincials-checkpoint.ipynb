```python
teams = []

for i in tba.event_teams(event):
    teams += [int(i['key'][-len(i['key'])+3:])]
    
teams
```

```python
df2 = pd.DataFrame(columns=['teamNumber','match', 'climb'])
for t in teams:
    print(t)
    max_week = 0
    for e in tba.team_events(t, 2025):
        if e['week'] is not None:
            max_week = max(max_week, e['week'])
    for e in tba.team_events(t, 2025):
        if e['week'] == max_week:
            for m in tba.team_matches(t, e['key']):
                blue_teams = m['alliances']['blue']['team_keys']
                red_teams = m['alliances']['red']['team_keys']
                
                team_string = 'frc' + str(t)
                
                if team_string in blue_teams:
                    if team_string == blue_teams[0]:
                        end_status = m['score_breakdown']['blue']['endGameRobot1']
                    elif team_string == blue_teams[1]:
                        end_status = m['score_breakdown']['blue']['endGameRobot2']
                    elif team_string == blue_teams[2]:
                        end_status = m['score_breakdown']['blue']['endGameRobot3']
                elif team_string in red_teams:
                    if team_string == red_teams[0]:
                        end_status = m['score_breakdown']['red']['endGameRobot1']
                    elif team_string == red_teams[1]:
                        end_status = m['score_breakdown']['red']['endGameRobot2']
                    elif team_string == red_teams[2]:
                        end_status = m['score_breakdown']['red']['endGameRobot3']
                
                new_row = [t, m['key'], end_status]
                df2.loc[len(df2)] = new_row
```

```python
dict_ = sb.get_team_year(teams[0], 2025)['epa']['breakdown']
dict_['teamNumber'] = teams[0]
df3 = pd.DataFrame(dict_, index=[dict_['teamNumber']])

for t in teams[1:]:
    print(t)
    new_dict = sb.get_team_year(t, 2025)['epa']['breakdown']
    new_dict['teamNumber'] = t
    df3 = pd.concat([df3, pd.DataFrame(new_dict, index=[new_dict['teamNumber']])])
```

```python
print('Matches with at least 1 L1')
for e in events:
    count_ = 0
    score_ = 0
    for i in tba.event_matches(e):
        if i['comp_level'] == 'qm':
            blue = 1 if i['score_breakdown']['blue']['teleopReef']['trough'] + i['score_breakdown']['blue']['autoReef']['trough'] > 0 else 0
            red = 1 if i['score_breakdown']['red']['teleopReef']['trough'] + i['score_breakdown']['red']['autoReef']['trough'] > 0 else 0
            count_ += 2
            score_ += blue + red
    print(score_, count_)
```

```python
print('Matches decided by fouls')
for e in events:
    count_ = 0
    score_ = 0
    for i in tba.event_matches(e):
        if i['comp_level'] == 'qm':
            if i['alliances']['red']['score'] > i['alliances']['blue']['score']:
                if i['alliances']['red']['score'] - i['score_breakdown']['red']['foulPoints'] <= i['alliances']['blue']['score'] - i['score_breakdown']['blue']['foulPoints']:
                    score_ += 1
            elif i['alliances']['red']['score'] < i['alliances']['blue']['score']:
                if i['alliances']['red']['score'] - i['score_breakdown']['red']['foulPoints'] >= i['alliances']['blue']['score'] - i['score_breakdown']['blue']['foulPoints']:
                    score_ += 1
            else:
                if i['score_breakdown']['red']['foulPoints'] != i['score_breakdown']['blue']['foulPoints']:
                    score_ += 1
            count_ += 1
                
    print(e, score_, count_)
```
