```python
def uncommon_data(list1, list2):
    result = True
  
    # traverse in the 1st list
    for x in list1:
  
        # traverse in the 2nd list
        for y in list2:
    
            # if one common
            if x == y:
                result = False
                return result 
                  
    return result
```

```python
wins = {}
total = {}

stackevents = ['2015abca',
 '2015onnb',
 '2015onto',
 '2015onto2',
 '2015onwa',
 '2015onwi',
 '2015qcmo']

canada_win_events = [x for x in all_canada_events if x not in stackevents]

for q in canada_win_events:
    test = []
    for m in tba.event_matches(q, 'simple'):
        red = [c for c in combinations(tuple(sorted(m['alliances']['red']['team_keys'])),2)]
        blue = [c for c in combinations(tuple(sorted(m['alliances']['blue']['team_keys'])),2)]
        alliances = red + blue
        for a in alliances:
            if a in total:
                total[a] += 1
            else:
                total[a] = 1
        if m['winning_alliance'] == 'red':
            for r in red:
                if r in wins:
                    wins[r] += 1
                else:
                    wins[r] = 1
        elif m['winning_alliance'] == 'blue':
            for r in blue:
                if r in wins:
                    wins[r] += 1
                else:
                    wins[r] = 1
        elif m['winning_alliance'] == '':
            for r in alliances:
                if r in wins:
                    wins[r] += 0.5
                else:
                    wins[r] = 0.5
        
```

```python
cawinsdf = pd.DataFrame(list(wins.items()),columns = ['Pairs','Wins']) 
camatchessdf = pd.DataFrame(list(total.items()),columns = ['Pairs','Matches']) 
```

```python
cadf = cawinsdf.merge(camatchessdf, on='Pairs',how='right').fillna(0)
```

```python
cadf['Winrate'] = cadf['Wins']/cadf['Matches']
cadf['Loses'] = cadf['Matches'] - cadf['Wins']
```

```python
cadf_work = cadf[cadf['Matches'] > 3].sort_values(['Matches','Wins'], ignore_index=True, ascending=False)#.sort_values(['Winrate'], ignore_index=True)
```

```python
cadf_output =  cadf_work[pd.Series(cadf_work['Pairs']).apply(lambda x: 'frc1114' in x)].reset_index()
```
